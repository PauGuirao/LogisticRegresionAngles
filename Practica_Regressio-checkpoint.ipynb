```python
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn import datasets
import math
import csv

with open('C:/Users/guira/Downloads/triangles.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    headers = next(spamreader)
    x_E = np.array([])
    y_E = np.array([])
    x_I = np.array([])
    y_I = np.array([])
    for row in spamreader:
        tipus = row[2];
        if(int(tipus) == 1):
            x_E = np.append(x_E,float(row[0]))
            y_E = np.append(y_E,float(row[1]))
        else:
            x_I = np.append(x_I,float(row[0]))
            y_I = np.append(y_I,float(row[1]))
plt.scatter(x_E, y_E)
plt.scatter(x_I, y_I)
plt.show()
```
